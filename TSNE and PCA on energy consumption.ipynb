```python
from sklearn.datasets import make_regression
```

```python
X,y= make_regression(n_samples=443,n_features=1440,n_informative=1,n_targets=1,noise=20,random_state=13)
```

```python
model=GaussianRandomProjection(eps=0.9)
x_new=model.fit_transform(X,y)
x_new
```

```python
x_new.shape
```

```python
km=KMeans(n_clusters=4)
km
```

```python
x_pred=km.fit_predict(x_new)
x_pred
```

```python
silhouette_score(x_new,x_pred)
```

```python
plt.scatter(x_new[x_pred==0,0],x_new[x_pred==0,1],color='blue')
plt.scatter(x_new[x_pred==1,0],x_new[x_pred==1,1],color='green')
plt.scatter(x_new[x_pred==2,0],x_new[x_pred==2,1],color='red')
plt.scatter(x_new[x_pred==3,0],x_new[x_pred==3,1],color='yellow')
plt.savefig('random prejection')
```
